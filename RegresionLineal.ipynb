```python
import pandas as pd # Manipulación de dataframe
import seaborn as sns #Libreria visualización de datos
import matplotlib.pyplot as plt #Libreria permite creación de gráficos
```

```python
data = pd.read_csv('car_details_v4.csv')
data.head(10) #Dataframe general
```

```python
data.describe() #Descripción estadístico
```

```python
sns.pairplot(data)
plt.show() #Visualizar relaciones
```

```python
sns.boxplot(data=data) # Coprobar valores atípicos
plt.show() 
```

```python
#Tratar valores faltantes
#data.fillna(data.mean(),inplace=True) #Rellenar
data.dropna(inplace=True) #Eliminar filas faltantes
data = pd.get_dummies(data,drop_first=True) #Transformar columnas
data.head()
```

```python
from sklearn.feature_selection import SelectKBest, f_regression # sklearn es una libreria de aprendizaje
X = data.drop('Price', axis=1)
y = data['Price']

selector = SelectKBest(score_func=f_regression,k=10) #Seleccionar mejores características
X_new = selector.fit_transform(X,y)

selected_features = selector.get_support(indices=True) #Verificar características
X.columns[selected_features]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_new,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV

model = LinearRegression() #Configurar y entrenar el modelo
model.fit(X_train,y_train)

y_pred = model.predict(X_test) #Predicciones
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print(f'Mean Squard Error:  {mse}')
print(f'R^2 Score:          {r2}')
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel('Valores Reales')
plt.ylabel('Predicciones')
plt.title('Predicciones Vs Valores Reales')
plt.show()

residual = y_test - y_pred
sns.histplot(residual, kde=True)
plt.title('Distribución de Residuos')
plt.show()
```
